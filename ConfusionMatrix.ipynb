```python
from __future__ import division
import pandas as pd
import csv
import numpy as np
import os
```

```python
gt = pd.read_csv(os.getcwd() + "/GTResults.csv")
at = pd.read_csv(os.getcwd() + "/SegResults.csv")
```

```python
gt['combined'] = gt.apply(lambda x: [x.start_GT, x.end_GT], axis=1)
at['combined'] = at.apply(lambda x: [[x.start, x.end]], axis=1)
at['combined'] = at.apply(lambda x: [x.start, x.end], axis=1)
```

```python
seg_result = list(at['combined'])
seg_result = np.array(seg_result)
```

```python
A = 0
B = 0
C = 0
gt_copy = gt.loc[gt.Speaker != "0"].reset_index(drop = 1)
missed_id = []
seg_copy = seg_result.copy()
for idx, ground_truth in gt_copy.iterrows():
    found = np.where((seg_copy[:,0] >= ground_truth.start_GT - 1) & (seg_copy[:, 1] <= ground_truth.end_GT + 1))[0]
    if found.shape[0] == 0:
        B += 1
        missed_id.append(idx)
    else:
        A += 1
        seg_copy[found] = np.array([-1, -1])
C = sum(seg_copy[:, 0] != -1)
```

```python
FDR = C/(A+B+C)
MDR = B/(A+B)
PRC = A/(A+C)
RCC = A/(A+B)
F = 2 * PRC * RCC /(PRC+RCC)
```
